```python
import random, shutil
from pathlib import Path

# 설정
# 1. 원본 사진들이 있는 곳 (아까 이름 바꾼 폴더)
SELECTED_DIR = Path(r"C:\gachikium\dataset\train") 

# 2. 새로 나눌 폴더 이름 (기존 train과 겹치지 않게 이름을 살짝 바꿉니다)
TRAIN_DIR    = Path(r"C:\gachikium\dataset\train_final")
TEST_DIR     = Path(r"C:\gachikium\dataset\test_final")

# 3. 중요! 원본 보존을 위해 반드시 복사 모드로 하세요.
COPY_MODE = True

TRAIN_RATIO = 0.8
SEED = 42
COPY_MODE = True         # True=복사, False=이동
CLEAR_OUTPUT = True      # train/test 폴더를 비우고 다시 생성

IMG_EXTS = {".jpg", ".jpeg", ".png", ".webp", ".bmp"}

random.seed(SEED)

def copy_or_move(src: Path, dst: Path):
    dst.parent.mkdir(parents=True, exist_ok=True)
    if COPY_MODE:
        shutil.copy2(src, dst)   # 동일 파일명 있으면 덮어씀
    else:
        shutil.move(str(src), str(dst))

def clear_dir(path: Path):
    if path.exists():
        shutil.rmtree(path)
    path.mkdir(parents=True, exist_ok=True)

def split_train_test(paths, train_ratio=0.8):
    paths = list(paths)
    random.shuffle(paths)
    n = len(paths)
    n_train = int(n * train_ratio)
    if n >= 2:
        n_train = min(max(n_train, 1), n - 1)  # 최소 1장씩은 배정
    return paths[:n_train], paths[n_train:]

# 실행
if not SELECTED_DIR.exists():
    raise FileNotFoundError(f"Not found: {SELECTED_DIR.resolve()}")

if CLEAR_OUTPUT:
    clear_dir(TRAIN_DIR)
    clear_dir(TEST_DIR)

for cls_dir in [d for d in SELECTED_DIR.iterdir() if d.is_dir()]:
    imgs = [p for p in cls_dir.iterdir() if p.is_file() and p.suffix.lower() in IMG_EXTS]
    if not imgs:
        print(f"[WARN] no images: {cls_dir.name}")
        continue

    train_list, test_list = split_train_test(imgs, TRAIN_RATIO)

    for src in train_list:
        copy_or_move(src, TRAIN_DIR / cls_dir.name / src.name)

    for src in test_list:
        copy_or_move(src, TEST_DIR / cls_dir.name / src.name)

    print(f"split {cls_dir.name}: total={len(imgs)}, train={len(train_list)}, test={len(test_list)}")

print("\nDone!")
print("Train:", TRAIN_DIR.resolve())
print("Test :", TEST_DIR.resolve())
```

# 1. 파일 이동

```python
import os, shutil # os: 경로/폴더/파일 목록 다루기, shutil: 파일 이동(move) 같은 작업
import random #랜덤 셔플(섞기) 위해 사용 

def dataset_split(query, train_cnt, seed=42): # query: 클래스(폴더) 이름, train_cnt: train 폴더에 남길 이미지 개수 (나머지는 test로 이동), seed: 섞는 순서를 고정해서 매번 같은 결과가 나오게 함(재현성)
    # 데이터셋 상위 디렉토리 (dataset 폴더를 기준으로 train/test를 관리)
    base_dir = './dataset'

    # 클래스별 train/test 폴더 경로 만들기 
    # 예../dataset/train/강아지상, ./dataset/test/강아지상
    train_dir = os.path.join(base_dir, 'train', query)  
    test_dir = os.path.join(base_dir, 'test', query) 

    # 폴더가 없으면 생성 (이미 있으면 그냥 넘어감)
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # train_dir 자체가 없다면 함수 종료
    # 보통은 위에서 makedirs로 만들었지만 안전장치로 둠. 
    if not os.path.exists(train_dir):
        return
    
    # train 폴더 안의 파일 목록 싹 가져오기
    files = os.listdir(train_dir)

    # 혹시 폴더/숨김파일 등 섞여 있을 수 있으니
    # '진짜 파일'만 남기도록 필터링
    # (os.path.isfile(..)가 True 인 애들만)
    # 폴더/숨김 파일 제외(안전)
    files = [
        f for f in files 
        if os.path.isfile(os.path.join(train_dir, f))] #파일명을 실제 경로로 변환시켜줌 "./dataset/train/강아지상/image_1.jpg"이걸 만들어주는 코드

    # 랜덤 셔플을 매번 동일하게 하려면 seed 고정 
    random.seed(seed)

    # 파일 순서를 랜덤하게 섞기
    # 이렇게 해야 특정 파일만 test로 몰리는 편향을 줄일 수 있음
    random.shuffle(files)

    # files를 앞에서 부터 순서대로 하나씩 꺼내면서 
    # i가 train_cnt 이상이면 test로 이동 
    for i, file_name in enumerate(files): # 파일에 번호를 매겨서 하나씩 추출
        # 현재 파일의 원래 위치(경로) 만들기
        # 예) ./dataset/train/강아지상/xxx.jpg
        src_path = os.path.join(train_dir, file_name)

        # 목적지 디렉토리 선택 로직
        # i < train_cnt -> train에 남김
        # i >= train_cnt -> test로 보냄 
        # 이거 안하면 파일 20개 , train_cnt = 16은 정상/ 파일 20개 vs train_cnt =40 은 실수야. 왜? 모든 파일이 train에 남게 되니까. 
        dst_dir = train_dir if i < train_cnt else test_dir # 개수에 

        # train에 남기는 건 굳이 move할 필요 없음
        # test로 가는 경우에만 move 수행 
        if dst_dir != train_dir:  # 시험 용으로 분류된 파일만 실제로 이동
            shutil.move(src_path, os.path.join(dst_dir, file_name)) # 개수에 맞춰 학습/시험 목적지 결정 

    # 결과 출력
    # train에 남긴 개수: train_cnt가 파일 개수보다 클 수 있으니 min 처리 
    # test로 간 개수 : 전체 - train_cnt(0미만이면 0처리)
    print(f"{query}: train={min(train_cnt, len(files))}, test={max(0, len(files)-train_cnt)}")


# 1) 10가지 동물상 리스트
animal_faces = [
    '강아지상', '고양이상', '사슴상', '토끼상', '꼬부기상',
    '곰상', '쿼카상', '쥐상', '도롱뇽상', '여우상'
]

# 2) 각 클래스가 20장이라면 → train 16 / test 4 추천
for face in animal_faces:
    print(f"--- {face} 데이터 분할 시작 ---")
    dataset_split(face, train_cnt=16, seed=42)
```

# 2. Crop & Resize

```python
import cv2, numpy as np
from pathlib import Path
from insightface.app import FaceAnalysis
from insightface.utils.face_align import norm_crop

# ===== 유니코드 경로 read/write =====
def imread_unicode(path: Path):
    data = np.fromfile(str(path), dtype=np.uint8)
    return cv2.imdecode(data, cv2.IMREAD_COLOR)

def imwrite_unicode_jpg(path: Path, img):
    path.parent.mkdir(parents=True, exist_ok=True)
    ok, buf = cv2.imencode(".jpg", img)
    if not ok:
        raise RuntimeError(f"imencode failed: {path}")
    buf.tofile(str(path))

def center_square_resize(img, size=224):
    h, w = img.shape[:2]
    s = min(h, w)
    y1 = (h - s) // 2
    x1 = (w - s) // 2
    crop = img[y1:y1+s, x1:x1+s]
    return cv2.resize(crop, (size, size), interpolation=cv2.INTER_AREA)

IMG_EXTS = {".jpg", ".jpeg", ".png", ".webp", ".bmp"}

# ===== 설정: 너 폴더로 바꿔 =====
TRAIN_IN  = Path(r"C:\gachikium\dataset\train")       # 이미 수동 split 한 train
TEST_IN   = Path(r"C:\gachikium\dataset\test")        # 이미 수동 split 한 test
TRAIN_OUT = Path(r"C:\gachikium\dataset\train_crop")  # 결과 저장
TEST_OUT  = Path(r"C:\gachikium\dataset\test_crop")   # 결과 저장

OUT_SIZE = 224  # ✅ ResNet이면 224 추천. (InsightFace는 112도 OK)

# ===== 얼굴 검출기 =====
app = FaceAnalysis(name="buffalo_l")
app.prepare(ctx_id=0, det_size=(1280, 1280), det_thresh=0.1)

def detect_best_face(app, img):
    # 1) 원본
    faces = app.get(img)
    if faces:
        face = max(faces, key=lambda f: (f.bbox[2]-f.bbox[0])*(f.bbox[3]-f.bbox[1]))
        return face, img

    # 2) padding
    h, w = img.shape[:2]
    pad = int(0.25 * max(h, w))
    img_pad = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_CONSTANT, value=(0, 0, 0))
    faces = app.get(img_pad)
    if faces:
        face = max(faces, key=lambda f: (f.bbox[2]-f.bbox[0])*(f.bbox[3]-f.bbox[1]))
        return face, img_pad

    # 3) upsample
    scale = 1.5
    img_up = cv2.resize(img, (int(w*scale), int(h*scale)), interpolation=cv2.INTER_CUBIC)
    faces = app.get(img_up)
    if faces:
        face = max(faces, key=lambda f: (f.bbox[2]-f.bbox[0])*(f.bbox[3]-f.bbox[1]))
        return face, img_up

    return None, img

def crop_face(img):
    face, used_img = detect_best_face(app, img)
    if face is not None and getattr(face, "kps", None) is not None:
        out = norm_crop(used_img, face.kps)  # 기본 112x112
        # 원하는 사이즈로 맞추기
        if OUT_SIZE != 112:
            out = cv2.resize(out, (OUT_SIZE, OUT_SIZE), interpolation=cv2.INTER_AREA)
        return out, True
    else:
        return center_square_resize(img, OUT_SIZE), False

def process_split(IN_DIR: Path, OUT_DIR: Path, split_name="train"):
    OUT_DIR.mkdir(parents=True, exist_ok=True)

    class_dirs = [d for d in IN_DIR.iterdir() if d.is_dir()]
    if not class_dirs:
        raise RuntimeError(f"No class folders found in: {IN_DIR}")

    total, ok_face, fallback = 0, 0, 0

    for cls_dir in sorted(class_dirs, key=lambda p: p.name):
        imgs = [p for p in cls_dir.iterdir() if p.is_file() and p.suffix.lower() in IMG_EXTS]
        if not imgs:
            print(f"[WARN] {split_name} {cls_dir.name}: no images")
            continue

        out_cls = OUT_DIR / cls_dir.name
        out_cls.mkdir(parents=True, exist_ok=True)

        for p in imgs:
            img = imread_unicode(p)
            if img is None:
                print(f"[WARN] unreadable: {p}")
                continue

            cropped, used_detector = crop_face(img)

            if used_detector:
                ok_face += 1
            else:
                fallback += 1

            # 같은 파일명 유지(원하면 여기서 규칙 rename도 가능)
            out_path = out_cls / (p.stem + ".jpg")
            imwrite_unicode_jpg(out_path, cropped)

            total += 1

        print(f"{split_name} {cls_dir.name}: {len(imgs)} processed")

    print(f"\n[{split_name}] total={total}, face_detected={ok_face}, fallback_center_crop={fallback}")
    print(f"Output: {OUT_DIR.resolve()}")

# ===== 실행 =====
process_split(TRAIN_IN, TRAIN_OUT, "train")
process_split(TEST_IN, TEST_OUT, "test")
```

# 3. 전처리와 증강

```python
from torchvision import transforms 

train_transforms = transforms.Compose([
    transforms.RandomResizeCrop(224, scale=(0.85, 1.0)) # 다양한 크롭/ 줌으로 일반화 
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.10),

    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])  
])

#  test transform
test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),  # 랜덤X
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
```

Dataset/ DataLoader 설계

```python
import torch
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

train_ds = ImageFolder("dataset/train", transform=train_transforms)
test_ds  = ImageFolder("dataset/test",  transform=test_transforms)

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True,  num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False, num_workers=2, pin_memory=True)

num_classes = len(train_ds.classes)
print("classes:", train_ds.classes)
print("num_classes:", num_classes)
```

# 4. 모델 설계 및 초기화

```python
import torch.nn as nn
from torchvision import models

device = "cuda" if torch.cuda.is_available() else "cpu"

model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

# backbone freeze
for p in model.parameters():
    p.requires_grad = False

# classifier 교체
in_features = model.fc.in_features
model.fc = nn.Linear(in_features, num_classes)

model = model.to(device)
```

Early Stopping 걸어서 epoch 너무 길게 끌고 가지 않음. 

```python
import torch
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=1e-3)  # 마지막 층만 학습

def train_one_epoch():
    model.train()
    total_loss, correct, total = 0.0, 0, 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)                 # (1) forward
        loss = criterion(outputs, labels)       # (2) loss 계산
        loss.backward()                         # (3) backward
        optimizer.step()                        # (4) update

        total_loss += loss.item() * inputs.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / total, correct / total

@torch.no_grad()
def evaluate():
    model.eval()
    total_loss, correct, total = 0.0, 0, 0

    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        outputs = model(inputs)
        loss = criterion(outputs, labels)

        total_loss += loss.item() * inputs.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / total, correct / total

# 1단계 학습
for epoch in range(1, 11):
    tr_loss, tr_acc = train_one_epoch()
    te_loss, te_acc = evaluate()
    print(f"[Stage1][{epoch:02d}] train loss {tr_loss:.4f} acc {tr_acc:.3f} | test loss {te_loss:.4f} acc {te_acc:.3f}")
```

- pretrained + freeze: 작은 데이터 과적합 방지

- RandomResizedCrop: 얼굴 크기/위치 다양성 보정(핵심 증강)

- test는 CenterCrop 고정: 평가 신뢰성 유지

- 2단계 미세조정: 필요할 때만, 아주 조금
